```python
missing_icao = ['KIWA', 'KBBG', 'KCRQ', 'KGPI', 'KSAW', 'KUNV', 'PGSN', 'KSGJ', 'KNYL']
missing_airports = airports[airports.IcaoIdenti.isin(missing_icao)].copy()
missing_airports
```

```python
missing_airports.LocationID.replace({'GSN': 'SPN',
                                     'SGJ': 'UST',
                                     'IWA': 'AZA',
                                     'CRQ': 'CLD',
                                     'NYL': 'YUM',
                                     'SAW': 'MQT',
                                     'BBG': 'BKG',
                                     'GPI': 'FCA',
                                     'UNV': 'SCE'}, inplace = True)
missing_airports
```

```python
airports_new = airports_new.append(missing_airports, ignore_index = True)
airports_new
```

```python
set(flights.Dest.unique()) - set(airports_new.LocationID)
```

```python
airports_clean = airports_new[['LocationID', 'StateName', 'State', 'FacilityNa', 'longitude_', 'latitude_d']].copy()
airports_clean
```

```python
airports_clean.StateName = airports_clean.StateName.str.title()
```

```python
airports_clean.FacilityNa = airports_clean.FacilityNa.str.title()
```

```python
airports_clean = airports_clean.assign(Airport = airports_clean.StateName + ', ' + airports_clean.State + 
                                       ': ' + airports_clean.FacilityNa)
```

```python
airports_clean.drop(columns = ['StateName', 'State', 'FacilityNa'], inplace = True)
```

```python
airports_clean
```

```python
airports_clean.to_csv('data/airports4.csv', encoding = 'utf-8', index = False)
```
